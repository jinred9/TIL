##### 1) 실습용 'titanic' 데이터셋

```python
import seaborn as sns
TD = sns.load_dataset('titanic')
TD.head()
```

##### 2) 'age'가 10살 이상이면서 20살 미만

```python
Filter_1 = (TD.age >= 10) & (TD.age < 20)
TD.loc[Filter_1, :].head()
```

##### 3) 'age'가 10살 미만이면서 'sex'가 여자

```python
Filter_2 = (TD.age < 10) & (TD.sex == 'female')
TD.loc[Filter_2, :].head()
```
