```python
print(df['Sepal_Length'].max())
print(df['Sepal_Length'].min())
print(df['Sepal_Length'].mean())
```

```python
print(df['Sepal_Length'].median())
print(df['Sepal_Length'].quantile())
```

```python
print(df['Sepal_Length'].quantile(q=0.25))
print(df['Sepal_Length'].quantile(q=0.5))
print(df['Sepal_Length'].quantile(q=0.75))
```

```python
df.iloc[:, :-1].isna().sum().sum()
```

```python
round(df.fillna(value={'Sepal_Width' : df['Sepal_Width'].mean()})['Sepal_Width'].std(), 3)
```

```python
round(df.fillna(value={'Sepal_Width' : df['Sepal_Width'].mean()})['Sepal_Width'].var(), 3)
```
