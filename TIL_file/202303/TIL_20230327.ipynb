```python
bike = pd.read_csv('../../ref_files/csv_file/bike.csv')
bike.head()
```

```python
bike['datetime'] = pd.to_datetime(bike['datetime'])
bike.info()
```

```python
bike['hour'] = bike['datetime'].dt.hour
```

```python
bike_s = bike[bike['season'] == 2]
bike_w = bike[bike['season'] == 4]
bike_sw = (bike_s.groupby('hour')['registered'].mean() - bike_w.groupby('hour')['registered'].mean()).abs().reset_index()
bike_sw
```

```python
bike_sw_max = bike_sw.max()
bike_sw_max
```

```python
bike_sw[bike_sw['registered'] ==  bike_sw_max]
```

```python
bike_s1 = bike[bike['season'] == 2].groupby('hour')['registered'].mean().reset_index()
bike_w1 = bike[bike['season'] == 4].groupby('hour')['registered'].mean().reset_index()
```

```python
bike_s1
```

```python
pd.merge(left=bike_s1, right=bike_w1, on='hour')
```
